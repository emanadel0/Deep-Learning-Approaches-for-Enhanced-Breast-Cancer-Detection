```python
# рџ“Њ 1) ШЄШ­Щ…ЩЉЩ„ Ш§Щ„Щ…ЩѓШЄШЁШ§ШЄ
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# рџ“Њ 2) ШЄШ­Щ…ЩЉЩ„ Ш§Щ„ШЇШ§ШЄШ§ (dataset Ш¬Ш§Щ‡ШІШ© Щ…Щ† keras)
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()

# Щ†Ш­Щ€Щ„Щ‡Ш§ binary (ШЁШі Щ„Щ„ШЄШ¬Ш±ШЁШ©)
train_labels = (train_labels > 4).astype(int)
test_labels = (test_labels > 4).astype(int)

# рџ“Њ 3) Normalization
train_images = train_images / 255.0
test_images = test_images / 255.0

# рџ“Њ 4) ШЁЩ†Ш§ШЎ Щ…Щ€ШЇЩЉЩ„ CNN ШЁШіЩЉШ·
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

# рџ“Њ 5) Compile
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# рџ“Њ 6) Training
history = model.fit(train_images, train_labels, epochs=5,
                    validation_data=(test_images, test_labels))

# рџ“Њ 7) Evaluation
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy:", test_acc)
```
